  Write a function replication that calculates the replicating portfolio   for an arbitrary claim in a one-period model. The parameters are 

    the initial bond price,
    the riskfree rate,
    the initial stock price,
    the up factor,
    the down factor,
    the payoff if the stock appreciates and
    the payoff if the stock depreciates.

Use np.linalg.solve and round all results to four decimals.

```python
import numpy as np

def replication(B0, r, S0, u, d, Cu, Cd):
    bond_future = B0 * (1 + r)
    A = np.array([
        [bond_future, S0 * u],
        [bond_future, S0 * d]
    ])
    b = np.array([Cu, Cd])
    x = np.linalg.solve(A, b)
    return np.round(x, 4)
```

```python
import numpy as np
import scipy.stats as sc
from scipy.stats import binom

def payoff(S0, K, flag):
    if flag.lower() == 'call':
        return np.maximum(S0 - K, 0)
    else:
        return np.maximum(K - S0, 0)
    
def calculate_price(S0, K, v, r, dt, n, flag):
    u = np.exp(v * np.sqrt(dt))
    d = 1/u
    qu = (np.exp(r * dt)-d)/(u-d)
    qd = 1 - qu
    a = np.arange(0, n + 1)
    ST = S0 * (u ** a) * (d ** (n - a))
    option_payoffs = payoff(ST, K, flag)
    probabilities = binom.pmf(a, n, qu)
    expected_payoff = np.sum(probabilities * option_payoffs)
    option_price = np.exp(-r * dt * n) * expected_payoff
    return np.round(option_price, 4)
print(calculate_price(100, 105, 0.33, 0.05, 1, 3, 'call'))
```

```python
import math

def cont_to_partial(r_cont, periods_per_year):
    """
    Converts a continuously compounded annual rate to a partial period rate.
    
    :param r_cont: float, continuously compounded annual rate (as a decimal, e.g., 0.05 for 5%)
    :param periods_per_year: int, number of periods in a year (e.g., 12 for monthly)
    :return: float, the partial period rate
    """
    t = 1 / periods_per_year
    return math.exp(r_cont * t) - 1

# Example: 5% continuous annual rate converted to a monthly rate
r_cont = 0.08
monthly_rate = cont_to_partial(r_cont, 12)

print(f"Continuous Rate: {r_cont * 100:.2f}%")
print(f"Equivalent Monthly Rate: {monthly_rate * 100:.4f}%")
```

    What is the value of a one-month European call option with a  strike price EUR ?


```python
S0 = 100    # Stock Price
u = 1.2     # Up factor
d = 0.8     # Down factor
r = 0.05    # Risk-free rate (5%)
K = 105     # Strike Price

def risk_neutral_one_period_call(S0, u, d, r, K):
    """
    Calculates the risk-neutral price of a European Call Option 
    using a one-period binomial framework.
    """
    # 1. Calculate Risk-Neutral Probabilities
    q_u = ((1 + r) - d) / (u - d)
    q_d = 1 - q_u
    # Check for no-arbitrage condition (d < 1+r < u)
    if not (d < (1 + r) < u):
        raise ValueError("Arbitrage exists! Ensure that d < 1+r < u.")
    
    # 2. Calculate terminal stock prices and option payoffs
    S_up = S0 * u
    S_down = S0 * d
    
    payoff_up = max(S_up - K, 0)
    payoff_down = max(S_down - K, 0)

    # 3. Discount expected value back to present value (t=0)
    expected_payoff = (q_u * payoff_up) + (q_d * payoff_down)
    option_price = expected_payoff / (1 + r)

    return option_price
price = risk_neutral_one_period_call(S0, u, d, r, K)   
print(price)
```
